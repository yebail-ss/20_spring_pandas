1. 对数据科学来说，表格数据的习惯是变数variables与观察observations
2. 在Pandas框框框的建构来说，就是可以把变数和观察用索引的方式使用
3. 框框框的字典取的是某个变数所有观察, 用.loc[]取數次观察的所有变数

框框 = pd.DataFrame ( {
